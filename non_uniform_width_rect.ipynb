```python
arr = [(1,0,1), (2,3,3), (2,7,4), (1,8,34), (1,78,8), (2,20,30), (2,12,17), (1,65,43), (2,19,17), (2,16,16), (1,2,2), (1,24,25)]
"""
(c,x,y) ==> (color, x_coord, y_coord)
1 ==> Blue
2 ==> Red
"""
blue = []
for i in range(len(arr)):
    if(arr[i][0]==1):
        blue.append(arr[i])

red = []
for i in range(len(arr)):
    if(arr[i][0]==2):
        red.append(arr[i])
```

```python
def top(lst):
    tp = lst[0]
    for i in range(len(lst)):
        if(lst[i][2]>tp[2]):
            tp = lst[i]
    return tp

def bottom(lst):
    bm = lst[0]
    for i in range(len(lst)):
        if(lst[i][2]<bm[2]):
            bm = lst[i]
    return bm

def left(lst):
    lf = lst[0]
    for i in range(len(lst)):
        if(lst[i][1]<lf[1]):
            lf = lst[i]
    return lf

def right(lst):
    rt = lst[0]
    for i in range(len(lst)):
        if(lst[i][1]>rt[1]):
            rt = lst[i]
    return rt
```

```python
b_out_top = top(blue)
b_out_bottom = bottom(blue)
b_out_left = left(blue)
b_out_right = right(blue)

b_out = [b_out_top, b_out_left, b_out_bottom, b_out_right]

print(b_out)
```
